```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
path = "/content/Boston.csv"
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(path)
```

```python
data.head()
```

```python
data.corr()
```

```python
msk = np.random.rand(len(data)) < 0.8
x = data[msk]
y = data[~msk]
```

```python
trainx = np.asanyarray(x[['crim','zn','age','tax']])
trainy = np.asanyarray(x[['tax']])
```

```python
testx = np.asanyarray(y[['crim','zn','age','tax']])
testy = np.asanyarray(y[['medv']])
```

```python
testx.shape
```

```python
testy.shape
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(trainx,trainy)
```

```python
y = model.predict(testx)
```

```python
from sklearn import metrics
```

```python
print(metrics.r2_score(y,testy))
```
