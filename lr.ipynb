```python
from sklearn import linear_model
import numpy as np
import pandas as pd
```

```python
path ="/boston.csv"
data = pd.read_csv(path)
```

```python
df = data[['crim','indus','rm','tax','ptratio','b','lstat','medv']]
```

```python
msk = np.random.rand(len(df)) < 0.9
train = df[msk]
test = df[~msk]
```

```python
train.head()
```

```python
test.head()
```

```python
train.describe()
```

```python
trainx = np.asanyarray(train[['crim','indus','rm','tax','ptratio','b','lstat']])
trainy = np.asanyarray(train[['medv']])
```

```python
testx = np.asanyarray(test[['crim','indus','rm','tax','ptratio','b','lstat']])
testy = np.asanyarray(test[['medv']])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(trainx,trainy)
```

```python
yhat = model.predict(testx)
```

```python
print("mean absolute error",np.mean(np.absolute(testy - yhat)))
```

```python
print("mean  error",np.mean(testy-yhat))
```

```python
from sklearn.metrics import r2_score
```

```python
print("r2 score",r2_score(yhat,testy))
```
