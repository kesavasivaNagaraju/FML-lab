```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))


```

```python
x = np.linspace(-10,10)
plt.plot(x,sigmoid(x))
plt.axis("tight")
plt.title("sigmoid function")
plt.show()
```

```python
def tanh(x):
  return np.tanh(x)
  
```

```python
x = np.linspace(-10,10)
plt.plot(x,tanh(x))
plt.axis("tight")
plt.title("tanh function")
plt.show()
```

```python
def relu(x):
  arr = []
  for i in x:
     if i < 0:
         arr.append(0)
     else:
         arr.append(i)
  return arr
```

```python
x = np.linspace(-10,10)
plt.plot(x,relu(x))
plt.axis('tight')
plt.title("relu ")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def RELU(x):
   x1=[]
   for i in x:
     if i<0:
        x1.append(0)
     else:
        x1.append(i)
   return x1
x = np.linspace(-10, 10)
plt.plot(x, RELU(x))
plt.axis('tight')
plt.title('Activation Function :ReLU')
plt.show()  
```

```python
def softmax(x):
  return np.exp(x)/np.sum(np.exp(x),axis=0)
```

```python
x = np.linspace(-10,10)
plt.plot(x,softmax(x))
plt.axis('tight')
plt.title("softmax activation function")
plt.show()
```
